##**Limpeza de dados**

Esta etapa é a mais importante e porque os dados precisam estar limpos antes de começar qualquer análise. Se os dados não forem limpos minuciosamente, os resultados serão imprecisos podendo causar erros na hora de tomar decisões estratégicas. A tradução de todo o dataset é importante já que nem todos os stakeholders falam inglês.

As etapas de limpeza serão:

- Renomear colunas;
- Remover espaços;
- Remover duplicatas;
- Padronizar maiúsculas e minúsculas;
- Mudar os tipos dos atributos se precisar;
- Tratar dados faltantes.

###**Tabela: olist_orders_dataset**

```python
# Mostrar as 5 primeiras linhas
df_orders.head()
```

```python
# Número de linhas e colunas
df_orders.shape
```

```python
# Nomes das variáveis
df_orders.columns
```

```python
# Informações sobre os tipos de dados
df_orders.info()
```

```python
# Criar uma cópia
df_orders_copy = df_orders.copy()
```

```python
# Remover espaços iniciais e finais de todas as linhas
df_orders_copy.columns = df_orders_copy.columns.str.strip()
```

```python
# Colunas da tabela
df_orders_copy.columns
```

```python
# Renomeando colunas
df_orders_copy.rename(columns=
 {'order_id':'ordem_id',
  'customer_id':'cliente_id',
  'order_status':'status_do_pedido',
  'order_purchase_timestamp':'data_da_compra',
  'order_approved_at':'data_da_aprovacao',
  'order_delivered_carrier_date':'data_de_envio',
  'order_delivered_customer_date': 'data_de_entrega',
  'order_estimated_delivery_date': 'previsao_de_entrega'}, inplace=True)
```

```python
# Como ficou
df_orders_copy.columns
```

```python
df_orders_copy.info()
```

**Renomear as colunas de texto**

**Coluna: status_do_pedido**

```python
# Quantas vezes aparece cada palavra
df_orders_copy['status_do_pedido'].value_counts()
```

```python
# Criando o dicionário para substituir os nomes
status = {'delivered': 'Entregue',
                'shipped': 'Enviado',
                'canceled': 'Cancelado',
                'unavailable': 'Indisponível',
                'invoiced': 'Faturado',
                'processing': 'Processando',
                'created': 'Criado',
                'approved': 'Aprovado'}
```

```python
# Substituir os atributos usando replace()
df_orders_copy['status_do_pedido'] = df_orders_copy['status_do_pedido'].replace(status)
```

```python
# Como ficou agora
df_orders_copy['status_do_pedido'].value_counts()
```

**Tratamento de dados faltantes**

```python
# Dados faltantes
df_orders_copy.isna().sum()
```

Como os dados faltantes são de datas, então provavelmente ficam dependentes do status do pedido. Por exemplo, não há como registrar o envio para a logística se o pagamento ainda não foi aprovado. Portanto, deixarei como está porque isso não vai interferir na análise.

**Tratamento de valores duplicados**

```python
# Há valores duplicados?
linhas_duplas = df_orders_copy[df_orders_copy.duplicated(['ordem_id'], keep=False)]
if not linhas_duplas.empty:
    print('Há valores duplicados:')
    display(linhas_duplas)
else:
    print('Não há valores duplicados.')
```

**Mudando o formato das colunas**

```python
# Tipo das colunas
df_orders_copy.info()
```

**Colunas de data**

```python
# 'data_da_compra'
df_orders_copy['data_da_compra'] = pd.to_datetime(df_orders_copy['data_da_compra'], format='mixed', dayfirst=True)
# data_da_aprovacao
df_orders_copy['data_da_aprovacao'] = pd.to_datetime(df_orders_copy['data_da_aprovacao'], format='mixed', dayfirst=True)
# envio_para_logistica
df_orders_copy['data_de_envio'] = pd.to_datetime(df_orders_copy['data_de_envio'], format='mixed', dayfirst=True)
# envio_para_cliente
df_orders_copy['data_de_entrega'] = pd.to_datetime(df_orders_copy['data_de_entrega'], format='mixed', dayfirst=True)
# previsao_de_entrega
df_orders_copy['previsao_de_entrega'] = pd.to_datetime(df_orders_copy['previsao_de_entrega'], format='mixed', dayfirst=True)
```

```python
# Como ficou
df_orders_copy.info()
```

As colunas de data que estavam como tipo 'object' (texto) agora estão em formato 'datetime64[ns]' (data e hora).

**Criando novas colunas**

As colunas abaixo serão criadas para calcular médias, totais, etc.

```python
# Processamento do pedido = data_de_envio - data_da_aprovacao
df_orders_copy['processamento_do_pedido'] = (df_orders_copy['data_de_envio'] - df_orders_copy['data_da_aprovacao']).dt.days
```

```python
# Dias para entregar = data_de_entrega - data_de_envio
df_orders_copy['dias_para_entregar'] = (df_orders_copy['data_de_entrega'] - df_orders_copy['data_de_envio']).dt.days
```

```python
# Atraso nas entregas = previsao_de_entrega - data_de_entrega
df_orders_copy['dias_de_atraso'] = (df_orders_copy['previsao_de_entrega'] - df_orders_copy['data_de_entrega']).dt.days
```

As colunas abaixo serão criadas para analisar sazonalidades, médias, totais, etc, filtrando por data.

```python
# Crie a coluna 'dia da compra'
df_orders_copy['dia_comp'] = df_orders_copy['data_da_compra'].dt.day
```

```python
# Crie a coluna 'mês da compra'
df_orders_copy['mes_comp'] = df_orders_copy['data_da_compra'].dt.month
```

```python
# Crie a coluna 'ano da compra'
df_orders_copy['ano_comp'] = df_orders_copy['data_da_compra'].dt.year
```

A coluna abaixo será criada para calcular a média de dias para aprovação dos pagamentos

```python
# Crie a coluna de dias para a aprovação do pagamento
df_orders_copy['dias_para_aprovacao'] = (df_orders_copy['data_da_aprovacao'] - df_orders_copy['data_da_compra']).dt.days
```
