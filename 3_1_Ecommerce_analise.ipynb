**Tabela: pedidos**
***
Esta é a tabela de fato, a mais importante porque aqui foram registrados os dias e horários exatos das vendas. Foi de fato o registro do momento exato em que todos os eventos ocorreram. As outras tabelas se conectam com esta através de uma chave estrangeira, ou seja, uma coluna igual a alguma coluna que tenha aqui também. Nesta tabela é a chave primária e nas outras são chaves estrangeiras. O relacionamento entre as tabelas aqui é de 1:n (um para muitos). Mas esta tabela não se conecta com todas, mas com a maioria das outras, que se conectam entre si também por chaves primárias e estrangeiras.
***
***(lista de colunas)***

**ordem_id:** identificador único de cada pedido ***(texto)***.

**cliente_id:** identificador único de cada cliente. Esse código é a chave primária nesta tabela, que é a chave estrangeira na tabela de clientes ***(texto)***.

**status_do_pedido:** situação do pedido, que pode ser entregue, enviado,cancelado, indisponível, faturado, processando, criado e aprovado ***(texto)***.

**data_da_compra:** data e hora exata que o cliente fez a compra ***(data e hora)***.

**data_da_aprovacao:** data e hora exata em que o pagamento foi aprovado ***(data e hora)***.

**data_de_envio:** data e hora exata em que o pedido foi despachado para a transportadora. ***(data e hora)***.

**data_de_entrega:** data e hora exata em que o cliente recebeu o pedido. ***(data e hora)***.

**previsao_de_entrega:** data e hora exata prevista para que o cliente receba o pedido. ***(data e hora)***.

***(Colunas novas criadas)***

**processamento_do_pedido:** quantos dias demorou desde a aprovação do pedido até o envio para a transportadora ***(número inteiro)***.

**dias_de_atraso:** A diferença entre a data de entrega estimada e a data de entrega real. Valores negativos indicam entrega antes do previsto, e valores positivos indicam atraso ***(número inteiro)***.

### Receitas

```python
# Mesclar as tabelas df_pedidos e df_itens_de_pedidos
mesclar = pd.merge(df_pedidos, df_itens_de_pedidos, how='left', on='ordem_id')
```

```python
#  Calcule a receita total
total_receita = mesclar['total_do_pedido'].sum()
total_receita
```

Dezessete milhões, seiscentos e quarenta e nove mil, seiscentos e quarenta e oito reais e quinze centavos.

```python
# Calcule a receita total dos pedidos que já foram entregues
receita_total_e = mesclar[(mesclar['status_do_pedido'] == 'Entregue')]['total_do_pedido'].sum()
receita_total_e
```

Dezessete milhões, cento e oitenta mil, duzentos e doze reais e noventa e cinco centavos.

```python
# Valores em aberto (pedidos ainda não entregues, em processamento ou cancelados)
total_receita - receita_total_e
```

Quatrocentos e sessenta e nove mil, quatrocentos e trinta e cinco reais e dezenove centavos.

```python
# Converter a coluna 'data_da_compra' para o formato data e hora
mesclar['data_da_compra'] = pd.to_datetime(mesclar['data_da_compra'])

# Calcule a receita total por ano somente dos pedidos já entregues
ano1 = mesclar[(mesclar['status_do_pedido'] == 'Entregue') & (mesclar['data_da_compra'].dt.year == 2016)]['total_do_pedido'].sum()
ano2 = mesclar[(mesclar['status_do_pedido'] == 'Entregue') & (mesclar['data_da_compra'].dt.year == 2017)]['total_do_pedido'].sum()
ano3 = mesclar[(mesclar['status_do_pedido'] == 'Entregue') & (mesclar['data_da_compra'].dt.year == 2018)]['total_do_pedido'].sum()

# Plotar os valores no gráfico de colunas
rotulos = ['2016', '2017', '2018']
valores = [ano1, ano2, ano3]

# Dados ordenados
dados_ordenados = sorted(zip(rotulos, valores), key=lambda x: x[1], reverse=False)
eixo_x, eixo_y = zip(*dados_ordenados)

fig, ax = plt.subplots(figsize=(10, 6))

# width=1.0 para o espaço entre as barras
# Adicionei edgecolor='white' para criar uma linha fina de separação visual
bar_container = ax.bar(eixo_x, eixo_y, color='#9E2A3A', width=0.9, edgecolor='black')

# Configurações de estilo que já definimos
ax.grid(False)
ax.set_title('Receitas no período', fontsize=16, weight='bold', color='#BE3D2A', y=1.05)
ax.set_ylabel('Quantidade em milhões', fontsize=12, weight='bold', color='#BE3D2A')
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
# Adjust y-limit dynamically based on the maximum value, or a reasonable fixed value if no data
# ax.set_ylim(0, max(valores) * 1.1 if valores else 100000)
ax.set_ylim(0, 15000000)

# Remover as bordas superior e direita
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Porcentagens
ax.bar_label(
    bar_container,
    fmt=lambda x: f'{(x/receita_total_e) * 100:.1f} %',
    fontweight='bold',
    padding=3
)

plt.show()
```
