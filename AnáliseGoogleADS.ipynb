Remover o cifrão e transformar a variável em numérica (float) para assim poder analisar.

```python
# Remova os cifrões da variável 'custo' e converta para numérico de forma robusta
df_copy['custo'] = pd.to_numeric(
    df_copy['custo'].astype(str)
    .str.replace('$', '', regex=False)
    .str.replace('₹', '', regex=False),
    errors='coerce'
)

# .str.replace('$', '', regex=False): O parâmetro regex=False é importante porque,
# em expressões regulares, o símbolo $ tem um significado especial (fim da linha).
# Dizendo que não é regex, o Pandas trata apenas como um texto comum.
```

## Perguntas de negócio (KPIs)

```python
# Calcular o CPC: custo / cliques
cpc = round(df_lim['custo'].sum() / df_lim['cliques'].sum(),2)
print(f'O custo médio por clique é de ${cpc} (um dólar e cinquenta e cinco centavos)')
```

```python
# Calcular o CPL: custo / num_leads
cpl = round(df_lim['custo'].sum() / df_lim['num_leads'].sum(),2)
print(f'O custo médio por lead é de ${cpl} (dez dólares e setenta e seis centavos)')
```

```python
# Calcular o CPA: custo / conversoes
cpa = round(df_lim['custo'].sum() / df_lim['conversoes'].sum(),2)
print(f'O custo médio por lead é de ${cpa} (trinta e três dólares e três centavos)')
```

```python
# Calcular o ROAS: val_vendas / custo
roas = round(df_lim['val_vendas'].sum() / df_lim['custo'].sum(),2)
print(f'A receita gerada para cada real gasto em anúncio é de ${roas} (seis dólares e noventa e seis centavos)')
```

## 2 - Impressões

Quantas vezes o anúncio é exibido para o usuário.

```python
# Número total de impressões
tot_i = round(df_lim['impressoes'].sum(),)
print(f'Total de impressões = {tot_i}')
print('(Nove milhões e seiscentos e trinta e seis mil e novecentos e oito impressões)')
```

```python
# Somar a quantidade de impressões dos 4 anúncios
cam_1 = df_lim[df_lim['nome_da_campanha'] == 'Curso de dados A']['impressoes'].sum()
cam_2 = df_lim[df_lim['nome_da_campanha'] == 'Curso de dados B']['impressoes'].sum()
cam_3 = df_lim[df_lim['nome_da_campanha'] == 'Curso de dados C']['impressoes'].sum()
cam_4 = df_lim[df_lim['nome_da_campanha'] == 'Curso de dados D']['impressoes'].sum()

# Plotar os valores dos cliques no gráfico de colunas
rotulos = ['Curso A', 'Curso B', 'Curso C', 'Curso D']
valores = [cam_1, cam_2, cam_3, cam_4]

# Dados ordenados
dados_ordenados = sorted(zip(rotulos, valores), key=lambda x: x[1], reverse=True)
eixo_x, eixo_y = zip(*dados_ordenados)

fig, ax = plt.subplots(figsize=(10, 6))

# width=1.0 para o espaço entre as barras
# Adicionei edgecolor='white' para criar uma linha fina de separação visual
bar_container = ax.bar(eixo_x, eixo_y, color='#9E2A3A', width=0.9, edgecolor='black')

# Configurações de estilo que já definimos
ax.grid(False)
ax.set_title('Impressões dos anúncios', fontsize=16, weight='bold', color='#BE3D2A', y=1.05)
ax.set_ylabel('Quantidade 0 à 4 milhões', fontsize=12, weight='bold', color='#BE3D2A')
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.set_ylim(0, 4000000)

# Porcentagens
ax.bar_label(
    bar_container,
    fmt=lambda x: f'{(x/tot_i) * 100:.1f} %',
    fontweight='bold',
    padding=3
)

plt.show()
```
